### HIMARS & ATACMS

Пара слов по распиаренной теме с передачей США хинзирам HIMARS. 

01.06.2022 Колин Каль - замминистра обороны США - заявил, что принято решение о передаче хохлам 4 HIMARS (РСЗО, по-нашему). На фронт вся эта красота должна поступить после 3-х недельного курса обучения артиллеристов. Передается хохлам именно РСЗО, а не ОТРК (ATACMS) дальностью до 70 км. 

https://breakingdefense.com/2022/06/us-to-provide-4-himars-systems-to-ukraine-in-latest-700m-arms-package/

Этот факт товарищ подчеркнул отдельно: *"Мы обсуждали данный вопрос с украинскими официальными лицами". Однако правительство США в конечном счете посчитало, что достаточно будет HIMARS с ракетами средней дальности (до 70 км) … Мы не считаем, что им нужны системы с большей дальностью, способные поражать цели на сотни и сотни километров"*.

Что можно сказать?

1. Если дядя не брешет, то хохлам, видимо, передадут 4 РСЗО дальностью до 70 км - это "серьезное" усиление, учитывая, что по заявлениям МО РФ, мы как раз по 3-4 РСЗО в день и уничтожаем. 
2. РСЗО, скорее всего, будет на колесной базе - M142 с транспортно-пусковым контейнером на 6 ракет.
3. Головная часть - это, вероятнее всего, M30 А1/А2 с кассетами и полезной нагрузкой в 296 кг.
4. M30 А1/А2 имеют одну поганую особенность - они оснащены инерциальной ГСН со спутниковой коррекцией, что существенно уменьшит КВО.

В целом же стоит четко разделять понятия **HIMARS и ATACMS**. 

**HIMARS и MLRS - это РСЗО**, аналогичное нашим Смерчам, Торнадо и иже с ними. Шасси всего 2 - М142 на колесном ходу и М270 на гусеничном. М270 несет 2 ТПК по 6 ракет, а М142 - 1 ТПК на 6 ракет. Дальности у ракет аналогичные нашим: от 20 до 70 км. Начинка - кассетная весом в 296 кг. Ракеты у них разные. Есть старые неуправляемые М26 с дальностью до 45 км. А есть М30 с ГСН.

**ATACMS - это ОТРК**. На базе М142 можно таскать 1 ракету типа Block 1. На M270 - 2 ракеты, в том числе и Block 1А.
Block 1 - это ракета с кассетной БЧ, полезной нагрузкой в 560 кг и дальностью до 165 км.
Block 1A - это ракета либо с кассетной БЧ, либо с унитарной, полезной нагрузкой в 247 кг и дальностью до 300 км.
